# Matplotlib xy축 편집

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 가져오기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

## 여백 지정하기

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel('X-Axis', labelpad=15, loc='right')
plt.ylabel('X-Axis', labelpad=20, loc='top')
plt.show()
```
