## 정렬, Aggreagation 함수, GroupBy 적용

### DataFrame, Series의 정렬 - sort_values( )
DataFrame과 Series의 정렬을 위해서는 sort_values( ) 메서드를 이용한다. sort_values()는 RDBMS SQL의 order by키워드와 유사하다. sort_values()의 주요 입력 파라미터는 by, ascending, inplace이다. by로 특정 칼럼을 입력하면 해당 칼럼으로 정렬을 수행한다. ascending=True로 설정하면 오름차순으로 정렬하며, ascending=False로 설정하면 내림차순으로 설정된다. 기본(default)는 ascendling=True이다. inplace도 default는 inplace=False이다.

```python
# DataFrame sorting
# 이름을 알파벳 순서로
# titan_df.sort_values(by=['Name']) 오름차순
titan_df.sort_values(by=['Name'], ascending=False)  # 이름으로 내림차순
titan_df.sort_values(by=['Name','Pclass'])[:15]  # 1.이름으로 오름차순 2.Pclass오름차순/ 15개만 출력
```

```python
# Series의 sorting
titan_df.Age.sort_values()
```

```python
# 'Age':'max', 'SibSp':'sum', 'Fare':'mean'
# 행과 볼 조건을 dict로
    
agg_cond = {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
titan_df.groupby(by='Pclass').agg(agg_cond)
```

```python
agg_cond = {'Age':[max], 'SibSp':'sum', 'Fare':'mean'}
titan_df.groupby(by='Pclass').agg(agg_cond)
```

## apply lambda 식으로 데이터 가공
판다스는 apply함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공한다. 판다스의 경우 칼럼에 일괄적으로 데이터 가공을 하는 것이 속도 면에서 더 빠르나 복잡한 데이터 가공이 필요할 경우엔 어쩔 수 없이 apply lambda를 이용한다. 먼저 lambda 식에 익숙하지 않을 분들을 위해 설명을 아래와 같이 하겠다.

```python
# 제곱함수
def get_square(a):
    return a**2
```

```python
print(get_square(3))
```

```python
# lambda로 표현
lambda_square = lambda x: x**2
```

```python
print(lambda_square(3))
```

```python
# map함수와 lambda의 결합
A = [1,2,3]
list(map(lambda_square, A))
```

```python
list(map(lambda x: x**2, A))
```

#### Q. baby:5세 미만, child:12세 미만, teen:19세 미만, student:25세 미만, young adult:35세 미만, adult:60세 미만, elderly:60 이상

```python
def get_cat(age):
    cat=''
    if age < 5:
        cat = 'baby'
    elif age < 12:
        cat = 'child'
    elif age < 19:
        cat = 'teen'
    elif age < 25:
        cat = 'student'
    elif age < 35:
        cat = 'young_adult'
    elif age < 60:
        cat = 'adult'
    else:
        cat = 'elderly'
    return cat
```

```python
titan_df.Age.apply(lambda x: get_cat(x))
```
